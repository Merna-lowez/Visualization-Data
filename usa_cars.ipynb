```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import squarify
```

```python
df=pd.read_csv(r'C:\Users\Dell\Desktop\USA_cars_datasets.csv')
df.head()
```

```python
#Pie chart
plt.figure(figsize = (10, 10))
df['brand'].value_counts().head(10).plot.pie()
plt.title('Cars')
plt.legend(loc='upper left')
plt.show()
```

```python
#Column histogram
plt.figure(figsize = (8, 5))
ax=plt.gca()
ax.hist(df['price'] ,  color='red' , alpha=0.9, edgecolor = 'black' )
plt.title('Range of prices')
plt.xlabel("price")
plt.ylabel('No. of cars')
plt.show()
```

```python
#Line histogram 
# density (std)
# kde line (mean,min,max,std)
# hist 
sns.distplot(df['price'], kde = True, hist=True )
plt.xlabel('price') 
plt.show()
```

```python
#bubble chart
plt.figure(figsize = (10,5))
plt.scatter('brand', 'price', s='id',alpha=0.5, data=df,edgecolors="grey")
plt.title('Chart title')
plt.xlabel('brand')
plt.ylabel('price') 
plt.xticks(rotation=90)
plt.show()
```

```python
#scatter chart
plt.figure(figsize = (10,5))
plt.scatter( df["brand"], df["price"],alpha=0.5)
plt.xlabel('brand')
plt.ylabel('price') 
plt.xticks(rotation=90)
plt.show()
```

```python
#tree chart 
plt.figure(figsize = (10,10))
labels = df['brand'].unique()
sizes = df['brand'].value_counts()
color = [ 'pink', 'grey', 'orange','yellow','aqua','green','red','purple','white','blue','brown']
squarify.plot(sizes,color=color,label=labels)
plt.show()
```

```python
#subplots, at least two combined
figure, axes = plt.subplots(2,figsize = (8,8))
ax=df.plot('brand', 'price',ax=axes[0])
ax.get_legend().remove()
ax.set_xlabel('brand')
ax.set_ylabel('price')

ax=df.plot('mileage', 'price',ax=axes[1])
ax.get_legend().remove()

plt.show()
```

```python
#Box plot, at least two data frames
plt.figure(figsize=(12,6))
sorted_nb = df.groupby(['brand'])['price'].median().sort_values()
sns.boxplot(x=df['brand'], y=df['price'],order=list(sorted_nb.index))
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize = (6,6))
data1 = pd.DataFrame(np.random.rand(17,3), columns=['A','B','C']).assign(Location=1)
data2 = pd.DataFrame(np.random.rand(17,3)+0.2, columns=['A','B','C']).assign(Location=2)
data3 = pd.DataFrame(np.random.rand(17,3)+0.4, columns=['A','B','C']).assign(Location=3)

cdf = pd.concat([data1, data2, data3])    
mdf = pd.melt(cdf, id_vars=['Location'], var_name=['Letter'])
ax = sns.boxplot(x="Location", y="value", hue="Letter", data=mdf)    
plt.show()
```
